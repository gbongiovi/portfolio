# **1. Data Understanding**

* UID: unique identifier ranging from 1 to 10000
* productID: consisting of a letter L, M, or H for low (50% of all products), medium (30%), and high (20%) as product quality variants and a variant-specific serial number
* air temperature [K]: generated using a random walk process later normalized to a standard deviation of 2 K around 300 K
* process temperature [K]: generated using a random walk process normalized to a standard deviation of 1 K, added to the air temperature plus 10 K.
* rotational speed [rpm]: calculated from powepower of 2860 W, overlaid with a normally distributed noise
* torque [Nm]: torque values are normally distributed around 40 Nm with an Ïƒ = 10 Nm and no negative values.
* tool wear [min]: The quality variants H/M/L add 5/3/2 minutes of tool wear to the used tool in the process. and a
'machine failure' label that indicates, whether the machine has failed in this particular data point for any of the following failure modes are true.
* Important : There are two Targets - Do not make the mistake of using one of them as feature, as it will lead to leakage.
* Target : Failure or Not
* Failure Type : Type of Failure

# **3. Modelling**

## **3.1. Dataset Selection**

```r
Train_list <- list(train_c, trainB_c, train_fs, trainB_fs, train_cnf,trainB_cnf,
                   train_nf, trainB_nf)
Test_list <- list(test_c, test_fs, test_cnf, test_nf)
```

### **KNN**

```r
performance_knn<- expand.grid(
  dataset=c("C","C_Balanced","FS","FS_Balanced","CNF","CNF_Balanced","NF",
            "NF_Balanced"),
  accuracy=NA,
  f1=NA,
  recall=NA,
  precision=NA,
  GMean_PR=NA,
  GMean_SS=NA
)

for (i in 1:length(Train_list)){
  j=ceiling(i/2)
  prediction_knn<-data.frame()

  set.seed(123+i)
  model_knn <- knn(train = subset(Train_list[[i]],select=-c(type, target,
                                                            failure_type)),
                   test = subset(Test_list[[j]],select=-c(type, target,
                                                          failure_type)),
                   cl = Train_list[[i]][["target"]],
                   k = j,
                   prob = TRUE)

  # Evaluating performance
  prediction_knn<-rbind(prediction_knn, data.frame(prediction=as.numeric(as.character(model_knn)), target=as.numeric(as.character(Test_list[[j]][["target"]]))))
  prediction_knn$prediction<-as.factor(prediction_knn$prediction)
  prediction_knn$target<-as.factor(prediction_knn$target)

  aux = confusionMatrix(prediction_knn$prediction,prediction_knn$target, positive = "1")
  performance_knn$accuracy[i]<-aux[["byClass"]][["Balanced Accuracy"]]
  performance_knn$f1[i]<-aux[["byClass"]][["F1"]]
  performance_knn$recall[i]<-aux[["byClass"]][["Recall"]]
  performance_knn$precision[i]<-aux[["byClass"]][["Precision"]]
  performance_knn$GMean_PR[i]<-sqrt(aux[["byClass"]][["Precision"]]*aux[["byClass"]][["Recall"]])
  # Measuring GMean_SS
  a=aux[["byClass"]][["Specificity"]]
  b=aux[["byClass"]][["Sensitivity"]]
  performance_knn$GMean_SS[i]<-sqrt(a*b)
}

performance_knn <- as.data.frame(lapply(performance_knn, function(x) {
  if(is.numeric(x)) {
    round(x, 4)
  } else {
    x
  }
}))

performance_knn
```

### **Logistic Regression**
